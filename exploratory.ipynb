# Ford Gobike Exploratory data analysis from year 2017-2019

## Introduction
This data set contains the trip data of Gobike from year 2017 till 2019. The data set contains attributes such as start station, end station, trip duration, user type, gender, etc. We have the data for the past 2 years for this product to analyse its performance and what steps can be taken to improve it by understanding the demographics of the user base.

```python
import pandas as pd
import numpy as np
import seaborn as sb 
import matplotlib.pyplot as plt
sb.set(color_codes=True)
%matplotlib inline
```

```python
#reading the data from the csv file
#The data has been cleaned by me earlier in separate notebooks and saved in a master file  
df = pd.read_csv('gobike_master_data.csv')
```

```python
df
```

```python
df.info()
```

```python
#creating a copy of the data frame to retain original data 
df_clean=df.copy()
```

- Converting required data fields to Category type for easier analysis

```python
#Converting attributes to category type and time attributes to datetime for easier manipulation
df_clean['user_type'] = df_clean.user_type.astype('category')
df_clean['member_gender'] = df_clean.member_gender.astype('category')
df_clean['start_time'] = pd.to_datetime(df_clean['start_time'])
df_clean['end_time'] = pd.to_datetime(df_clean['end_time'])
day_order = [0,1,2,3,4,5,6]
df_clean['day_of_week'] = df_clean.day_of_week.astype('category', ordered=True, categories = day_order)
```

```python
df_clean.info()
```

### Consumer comparison(Univariate)

- Whenever we look at a product such as Gobike we need to understand what and how many users are using it. In this data set we have subscribers and customers, which can be further segregated into their genders. First let's find out which group of user type has a higher count?

```python
base_color = sb.color_palette()[0]
sb.countplot(data=df_clean, x='user_type',color=base_color);
plt.xlabel('User type');
plt.ylabel('Count');
plt.title('User type distribution');
```

```python
#to find out the percentage of the subscribers and the customers 
dist = []
dist.append(df_clean.user_type.value_counts()[0]/(df_clean.user_type.value_counts()[0]+df_clean.user_type.value_counts())[1]*100)
dist.append(df_clean.user_type.value_counts()[1]/(df_clean.user_type.value_counts()[0]+df_clean.user_type.value_counts()[1])*100)
dist
```

- I have created a count plot to compare the number of different user types which is a Univariate plot ,i.e., only one variable is being used. We can see that the subscribers are 83% while the customers are at approximately 17%. This graph could also be plotted in the form of a pie chart which is also a good way to plot a univariate analysis.

```python
dist = []
dist.append(df_clean.user_type.value_counts()[0]/(df_clean.user_type.value_counts()[0]+df_clean.user_type.value_counts())[1]*100)
dist.append(df_clean.user_type.value_counts()[1]/(df_clean.user_type.value_counts()[0]+df_clean.user_type.value_counts()[1])*100)
df_clean.user_type.unique()
labels = ['Customer', 'Subsciber']
patches, texts = plt.pie(dist, labels = df_clean.user_type.unique(), startangle = 90, counterclock = False);
plt.axis('square');
plt.legend(patches, labels, loc="best");
plt.title('User type distribution');
```

- To better understand the demographics we need to segregate this data further. What is the distributon of the males, females and other genders?

```python
#segregating the subscribers by their genders(bivariate)
#countplot between subscibers and customers 
ax = sb.countplot(data = df_clean, x = 'user_type', hue = 'member_gender');
ax.legend(loc = 2, framealpha = 1); # upper left, no transparency
plt.xlabel('User type');
plt.ylabel('Count');
plt.title('User gender distribution');
```

### Comparing the trip duration of different user types(bivariate analysis)

- After looking at the distribution of the user types, we should also know the trip duration these two categories usually book? 

```python
#to find out the mean trip durations of the user types
cdur = df_clean.query('user_type == "Customer"').duration_sec.mean()
sdur = df_clean.query('user_type == "Subscriber"').duration_sec.mean()
cdur, sdur
```

- We see here that the mean trip duration of the customer is more than twice of what it is of the subscribers.This does not feel right. For this I will plot a box plot to compare the means between the two user types and also to understand the outliers in the data set.

```python
#boxplot between user_type and duration_sec
sb.boxplot(data=df_clean, x='user_type', y='duration_sec', color=base_color);
plt.xticks(rotation=90);
plt.xlabel('User Type');
plt.ylabel('Trip Duration');
plt.title('User type vs Trip Duration');
```

- We can see above that the outlier values are much larger than the general data distribution and hence we are not able to get a clearer idea of the data distribution. So, I need to filter out the dution_sec data with much lower value to understand the same 

```python
#taking out the trip duration which is less than 4000 secs to get a more better graph 
bike_div = df_clean.query('duration_sec < 4000')
```

```python
#boxplot between user_type and duration_sec with duration sec less than 4000
sb.boxplot(data=bike_div, x='user_type', y='duration_sec', color=base_color);
plt.xticks(rotation=90);
plt.xlabel('User Type');
plt.ylabel('Trip Duration');
plt.title('User type vs Trip Duration');
```

- We can clearly see here that the mean usage time of the Customer is more than the subscriber, but not as much as we saw in the data computation by just calculating the mean. Therefore, sometimes a visual representation of the data provides us with a much better idea about the same. The one thing that we can understand from this is that the current customers should be converted to subscribers for better product statistics and value.

### Trip duration comparision between different genders of the suer types (multivariate analysis)

- Since we know that the trip duration for the customers is more than that of the subscribers and we also know that the the number of male users are much more than any of the other genders, is it safe to assume that the trip duration of the male members is also more than the other genders? I will plot a barplot between the mean trip duration between the genders in different user types which is a type of Multivariate analysis

```python
ax = sb.barplot(data = bike_div, x = 'user_type', y = 'duration_sec', hue='member_gender')
ax.legend(loc = 8, ncol = 3, framealpha = 1, title = 'User Gender')
plt.xlabel('User Type');
plt.ylabel('Trip Duration');
plt.title('User type vs Trip Duration with Gender distribution');
```

- In the above graph we see that even though the number of trips by males is more, the trip duration is lower than other gender groups. This was kind of a paradox and a misconeption many would have made if they do not see this visualisation. So despite the fact that more number of males are using the gobike, their mean trip duration is the lowest when you compare with the trip duration of other genders. This means that if a drive to focus to get other gender groups is initiated it would result in more revenue generation and also diversifies the product a bit more.

### Trips comparison between days of the week (Univariate analysis)
- One importnt thing to note is the trend of the trips being booked over the week. For this I will be plotting a graph of the trips over the different days of the week to understand this trend. I will use a facet grid plot to see this trend. Since the two main gender groups in this case are male and female, I will be using only these two gender groups for this analysis.

```python
df_div = df_clean.query('member_gender == "Male" or member_gender == "Female"')
g = sb.FacetGrid(data = df_div, col = 'day_of_week', hue = 'member_gender', hue_order = ['Male','Female'], size=3);
g.map(sb.countplot, 'user_type').add_legend(title = 'User Gender');
g.set_titles('{col_name}');
g.set_axis_labels('User type','Count');
g.set_titles('User type with Day of week distribution');
ax = g.axes.flatten();
ax[0].set_title('Monday');
ax[1].set_title('Tuesday');
ax[2].set_title('Wednesday');
ax[3].set_title('Thursday');
ax[4].set_title('Friday');
ax[5].set_title('Saturday');
ax[6].set_title('Sunday');
```

- Above we can clearly see that the trips during the week days with a positive trend till Thursday and then shows a negative trend thursday onwards till sunday. So to increase the trips during the weekends there should be more offers and discounts to make the trend positive. We can also assume that people prefer to go to work using Gobike.

### Trip duration trend over time with bike share (bivariate analysis)
- What is the trip duration trend over time? Is the trip duration over time increasing or decreasing? We need to understand this aspect of the data as well to get an idea how well the user are engaged with Gobike. If there is a decreasing trip duration it may also eventually mean that people are losing interest or less revenue generated from the product.

```python
#taking the data where the gender is either male or female
df_mf = df_clean.query('member_gender == "Female" or member_gender == "Male"')
df_mf
```

```python
plt.scatter(data=df_mf, x= 'start_time', y = 'duration_sec', alpha = 1/15);
plt.xticks(rotation = 90);
plt.xlabel('Time (yyyy-mm)');
plt.ylabel('Trip Duration');
plt.title('Trip duration trend');
```

- This is a scatter plot to understand the trend between the trip duration with time. I have added transperancy to avoid overplotting of data so that the trend can be seen clearly with less confusion. We see here that there is a negative trend. The Trip duration has been dropping and we need to figure out the root cause for the same.
